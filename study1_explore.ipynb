### Poisson Regression Analysis

>(a) Create a subject-level data set (492 rows) with the following variables: the above
>seven predictors, the number of observations for the given subject (up to 18), and his/her
>number of future thoughts (i.e., number of observations for which future=1). (Hint: The
>aggregate() function or dplyr::group by() may be useful.) Use descriptive statistics
>and graphs to inspect these variables, and describe any missing data.

**Model**: Poisson Regression
- **Outcome**: `n_future` (count of future thoughts per person)
- **Predictors**: age, sex, O (Openness), C (Conscientiousness), E (Extraversion), A (Agreeableness), N (Neuroticism)

```python
subject, n_obs, n_futures = "Subject", "n_obs", "n_futures" 
predictors = ["age", "sex", "O", "C", "E", "A", "N"]
agg_dct = {k: (k, "first") for k in predictors}
agg_dct[n_obs] = (subject, "size")
agg_dct[n_futures] = ("time_3", "sum")
df_poisson  = df_clean.groupby(subject, as_index=False).agg(**agg_dct)
assert (df_poisson[n_futures] <= df_poisson[n_obs]).all()
print(f"Number of unique subjects: {df_poisson['Subject'].nunique()}")
df_poisson.head()
```

```python
df_poisson[['age', 'sex', 'E', 'A', 'C', 'N', 'O']].isna().sum()
```

```python
# I made the decision to drop subjects with missing data.
# Seems like theo only sensible thing to do here.

df_poisson_clean = df_poisson.dropna(
    subset=['age', 'sex', 'E', 'A', 'C', 'N', 'O']
)

len(df_poisson_clean), len(df_poisson)
```

```python
df_poisson_clean.to_csv(
    "datasources-time-travel/Study 1/ETT_ESM_Study1_Poisson_clean.csv", index=False)
```

```python
# Explore the outcome variable
print(df_poisson_clean[n_futures].describe())
print(df_poisson_clean[n_futures].value_counts().sort_index().head(20))


# Informal Check if variance equals mean (Poisson assumption)
mean_n_future = df_poisson_clean[n_futures].mean()
var_n_future = df_poisson_clean[n_futures].var()

print(f"\nPoisson assumption check:")
print(f"Mean: {mean_n_future:.2f}, Variance: {var_n_future:.2f}")
print(f"Variance/Mean ratio: {var_n_future / mean_n_future:.2f}")
if var_n_future > mean_n_future * 1.5:
    print("Overdispersion detected")
else:
    print("Variance ≈ Mean (Poisson assumption met)")
```

```python
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.api as sm

model = smf.glm(
    formula=f"{n_futures} ~ age + sex + E + A + C + N + O",
    data=df_poisson_clean,
    family=sm.families.Poisson(),
    offset=np.log(df_poisson_clean[n_obs])
).fit()


# Dispersion statistic
dispersion = model.deviance / model.df_resid
print(f"Dispersion statistic: {dispersion:.2f}")
print("Overdispersion detected" if dispersion > 1.5 else "No strong overdispersion")
```

```python
quasi_model = smf.glm(
    formula=f"{n_futures} ~ age + sex + E + A + C + N + O",
    data=df_poisson_clean,
    family=sm.families.Poisson(),
    offset=np.log(df_poisson_clean[n_obs])
).fit(scale=dispersion)
model.params.compare(quasi_model.params)
quasi_model.summary()
```

```python
quasi_dispersion = quasi_model.deviance / quasi_model.df_resid
print(f"QuasiPoisson Dispersion statistic: {quasi_dispersion:.2f}")
print("Overdispersion detected" if quasi_dispersion >
      1.5 else "No strong overdispersion")
```
